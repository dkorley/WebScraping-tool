```python
data = []
for i in range(50):
    url = f'https://quotes.toscrape.com/page/{i}/'
    r = requests.get(url)
    if r.status_code == 200:

        soup = BeautifulSoup(r.content, 'html.parser')
        quotes = soup.find_all('div', class_='quote')

        
        for quote in quotes:
            text = quote.find('span', class_='text').text
            author = quote.find('small', class_='author').text
            data.append({'Quote': text, 
                         'Author': author})

df = pd.DataFrame(data)
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')
tables = soup.findAll('table', class_='wikitable sortable sticky-header sort-under mw-datatable col2left col6left')
data = []

for table in tables:
    rows = table.find_all('tr')
    for row in rows:
        row_data = []
        cells = row.find_all(['th','td'])
        for cell in cells:
            row_data.append(cell.get_text(strip=True))
        #if row_data:
        data.append(row_data)

df = pd.DataFrame(data)
df.to_csv('countries_population.csv', index=False, header=False)
print("Data saved to 'countries_population.csv'")
```

```python
url = "https://en.wikipedia.org/wiki/World_population"
r = requests.get(url)
soup = BeautifulSoup(r.content, 'lxml')
table = soup.find('table', class_="wikitable sortable")
titles = []
headers = table.find_all("th")
for header in headers:
    title = header.text.strip()
    titles.append(title)
df = pd.DataFrame(columns=titles)

row_data = []
rows = table.find_all("tr")
for row in rows:
    detail = row.find_all("td")
    if detail:
        row_data.append([info.text.strip() for info in detail])

for idx, rw in enumerate(row_data):
    df.loc[idx] = rw
df
```

```python
url = "https://en.wikipedia.org/wiki/The_World's_Billionaires"
re = requests.get(url)
soup = BeautifulSoup(re.text, "lxml")

table = soup.find("table", class_="wikitable sortable")
titles = []
headers = table.findAll("th")
for header in headers:
    titles.append(header.text.strip())
df1 = pd.DataFrame(columns=titles)

row_data = []
details = table.find_all("tr")
for detail in details:
    rows = [row.text.strip() for row in detail.findAll("td")]
    if rows:  #to get rid of empty list
        row_data.append(rows)
for idx, row in enumerate(row_data):
    df1.loc[idx] = row

df1
```
